# Task 2

When node $z$ in TREE-DELETE has two children, we could choose node $y$ as
its predecessor rather than its successor. What other changes to TREE-DELETE
would be necessary if we did so? Some have argued that a fair strategy, giving
equal priority to predecessor and successor, yields better empirical performance.
How might TREE-DELETE be changed to implement such a fair strategy?

    1. TREE-DELETE(T, z)
    2.    if z.left == NIL
    3.        TRANSPLANT(T, z, z.right)
    4.    elseif z.right == NIL
    5.        TRANSPLANT(T, z, z.left)
    6.    else y = TREE-MINIMUM(z.left)
    7.        if y.p != z
    8.            TRANSPLANT(T, y, y.left)
    9.            y.left = z.left
    10.           y.left.p = y
    11.       TRANSPLANT(T, z, y)
    12.       y.right = z.right
    13.       y.right.p = y

5-րդ տողը գտնում է տարրի հաջորդը (Succesor) , իսկ եթե TREE-MAXIMUM-ը փոխենք TREE-MINIMUM կգտնի նախորդը (Predecesor)։ Այս մոտեցումը արդար դարձնելու համար, կարող ենք առաջին անգամ կանչել TREE-MAXIMUM-ով հետո TREE-MINIMUM-ով և այդպես շարունակ, կամ պարզապես պատահական կերպով ընտրենք մեթոդներից մեկը։

 # Task 3

Suppose that we have numbers between 1 and 1000 in a binary search tree, and we
want to search for the number 363. Which of the following sequences could not be
the sequence of nodes examined?

    a. 2, 252, 401, 398, 330, 344, 397, 363.
    b. 924, 220, 911, 244, 898, 258, 362, 363.
    c. 925, 202, 911, 240, 912, 245, 363.
    d. 2, 399, 387, 219, 266, 382, 381, 278, 363.
    e. 935, 278, 347, 621, 299, 392, 358, 363.

c) 912 > 911 այնինչ 911-ից ձախ բոլորը պետք է իրենից փոքր լենել

d) 299 < 347 այնինչ 347-ից աջ բոլորը պետք է իրենից մեծ լենել
